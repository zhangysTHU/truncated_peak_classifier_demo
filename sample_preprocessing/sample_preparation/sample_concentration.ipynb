```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Cov_Mod_df = pd.read_csv("/BioII/lulab_b/huangkeyun/zhangys/alkb-seq/resources/Cov_Mod.txt",sep = '\t')
Cov_Mod_df['last_part'] = Cov_Mod_df['sample'].str.split('|').str[-1]
Cov_Mod_df['label'] = Cov_Mod_df['last_part'].map({'same': 0, 'diff': 1})
Cov_Mod_df
```

```python
samples_df = Cov_Mod_df.iloc[:, 0:-1-1]
samples_df
```

```python
sequence_df = pd.read_csv("/BioII/lulab_b/huangkeyun/zhangys/alkb-seq/resources/seq_struc.csv", names = ['sample', 'score1', 'score2', 'struc', 'sequence'])
sequence_df
#Cov_Mod_df['sequence']
```

```python
# 检查 'Sequence' 列是否包含 'N'，并删除包含 'N' 的行
original_row_count = len(sequence_df)  # 原始行数
localization_df = sequence_df[~sequence_df['sequence'].str.contains('N')]  # 保留不包含 'N' 的行
new_row_count = len(sequence_df)  # 处理后行数

# 计算删除的行数
deleted_rows = original_row_count - new_row_count

# 打印删除的行数
print(f"删除了 {deleted_rows} 行")
```

```python
# 创建一个字典，将DNA碱基映射到one-hot编码
base_to_onehot = {
    'A': [1, 0, 0, 0],  # A -> [1, 0, 0, 0]
    'T': [0, 1, 0, 0],  # T -> [0, 1, 0, 0]
    'C': [0, 0, 1, 0],  # C -> [0, 0, 1, 0]
    'G': [0, 0, 0, 1]   # G -> [0, 0, 0, 1]
}

# 函数：将DNA序列转化为one-hot编码的展平向量
def sequence_to_onehot(sequence):
    onehot = []
    for base in sequence:
        onehot.extend(base_to_onehot.get(base, [0, 0, 0, 0]))  # 获取one-hot编码
    return onehot

# 将每个DNA序列转换为one-hot编码并展平
sequence_df['onehot_sequence'] = sequence_df['sequence'].apply(sequence_to_onehot)

```

```python
sequence_df = pd.concat([samples_df, sequence_df['onehot_sequence']], axis=1)
sequence_df['onehot_sequence'] = sequence_df['onehot_sequence'].apply(lambda x: ''.join(map(str, x)))
sequence_df
```

```python
new_df = sequence_df.copy()

# 函数：对每个onehot_sequence字符串进行重新排序
def reorder_onehot_sequence(sequence):
    # 将字符串转化为列表，按4个数字一组进行切割
    chunks = [sequence[i:i+4] for i in range(0, len(sequence), 4)]
    
    # 创建新的四个部分，分别对应每组的第一、第二、第三、第四个数字
    part1 = ''.join([chunk[0] for chunk in chunks])  # 每组的第一个数字
    part2 = ''.join([chunk[1] for chunk in chunks])  # 每组的第二个数字
    part3 = ''.join([chunk[2] for chunk in chunks])  # 每组的第三个数字
    part4 = ''.join([chunk[3] for chunk in chunks])  # 每组的第四个数字
    
    # 合并四部分并返回新的字符串
    return part1 + part2 + part3 + part4

# 对每一行的 'onehot_sequence' 列进行重新排序
new_df['onehot_sequence'] = new_df['onehot_sequence'].apply(reorder_onehot_sequence)
new_df[['seq' + str(i+1) for i in range(160)]] = new_df['onehot_sequence'].apply(lambda x: pd.Series(list(x)))
new_df = new_df.drop(columns=['onehot_sequence'])
```

```python
new_df.to_csv('/BioII/lulab_b/huangkeyun/zhangys/alkb-seq/resources/sample_prepared.csv', sep = ',', index = None)
```
