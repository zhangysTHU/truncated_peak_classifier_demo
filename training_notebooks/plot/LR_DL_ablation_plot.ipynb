```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 读取DL消融结果CSV文件
file_path = '/BioII/lulab_b/huangkeyun/zhangys/alkb-seq/ML_models/eight_sample_11features_test/5fold_features_ablation/cross_ablation_posencoded_results.csv'
data = pd.read_csv(file_path)

# 提取数据
features_num = data['features_num']
mean_accuracy = data['mean_accuracy']
accuracy_variance = data['accuracy_variance']
mean_auroc = data['mean_auroc']
auroc_variance = data['auroc_variance']
mean_F1 = data['mean_F1']
F1_variance = data['F1_variance']
mean_aurp = data['mean_aurp']
aurp_variance = data['aurp_variance']

# 设置柱状图参数
bar_width = 0.2
index = np.arange(len(features_num))

# 绘制柱状图
fig, ax = plt.subplots(figsize=(12, 8))

bar1 = ax.bar(index, mean_accuracy, bar_width, yerr=accuracy_variance, label='Mean Accuracy')
bar2 = ax.bar(index + bar_width, mean_auroc, bar_width, yerr=auroc_variance, label='Mean AUROC')
bar3 = ax.bar(index + 2 * bar_width, mean_F1, bar_width, yerr=F1_variance, label='Mean F1')
bar4 = ax.bar(index + 3 * bar_width, mean_aurp, bar_width, yerr=aurp_variance, label='Mean AURP')

# 添加标签和标题
ax.set_xlabel('Features Num', fontsize=14)
ax.set_ylabel('Scores', fontsize=14)
ax.set_title('Grouped Bar Chart of DL reversed Ablation Results',  fontsize=16)
ax.set_xticks(index + 1.5 * bar_width)
ax.set_xticklabels(features_num)
ax.legend()

ax.set_ylim(0.45, 0.9)
# 显示图表
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
