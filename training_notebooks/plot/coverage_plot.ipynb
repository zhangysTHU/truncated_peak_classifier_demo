```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy.stats import ttest_ind
```

```python
# 计算每行的平均值
data_diff_avg1 = data_diff.iloc[:, :20].mean(axis=1)
data_same_avg1 = data_same.iloc[:, :20].mean(axis=1)

data_diff_avg2 = data_diff.iloc[:, 20:].mean(axis=1)
data_same_avg2 = data_same.iloc[:, 20:].mean(axis=1)
```

```python
data_diff_avg2
```

```python
# 定义显著性检验函数
def add_significance(ax, data1, data2, x_offset=0):
    # 进行t检验
    _, p_value = ttest_ind(data1, data2)
    
    # 确定显著性符号
    if p_value < 0.001:
        significance = '***'
    elif p_value < 0.01:
        significance = '**'
    elif p_value < 0.05:
        significance = '*'
    else:
        significance = 'ns'
    
    # 添加到图表上
    ax.text(x_offset, max(data1.max(), data2.max()) + 0.05, f'{significance} (p={p_value:.50f})', 
            ha='center', va='bottom', fontsize=10)
```

```python
# 绘制小提琴图
fig, axes = plt.subplots(1, 2, figsize=(15, 7), sharey=True)

# 第一组：data_diff和data_same的前20列平均值比较
sns.violinplot(data=[data_diff_avg1, data_same_avg1], ax=axes[0], palette=["skyblue", "lightgreen"])
axes[0].set_title('First 20 columns')
axes[0].set_xticklabels(['Data Diff', 'Data Same'])

# 添加显著性检验
# add_significance(axes[0], data_diff_avg1, data_same_avg1, x_offset=0)

# 第二组：data_diff和data_same的后20列平均值比较
sns.violinplot(data=[data_diff_avg2, data_same_avg2], ax=axes[1], palette=["skyblue", "lightgreen"])
axes[1].set_title('Last 20 columns')
axes[1].set_xticklabels(['Data Diff', 'Data Same'])

# 添加显著性检验
# add_significance(axes[1], data_diff_avg2, data_same_avg2, x_offset=0)

plt.tight_layout()
plt.show()
```

```python
# 绘制小提琴图
fig, axes = plt.subplots(1, 2, figsize=(15, 7), sharey=True)

# 第一组：data_diff和data_same的前20列平均值比较
parts = axes[0].violinplot([data_diff_avg1, data_same_avg1], showmedians=True, showmeans=False, showextrema=False)

# 设置颜色
for i, pc in enumerate(parts['bodies']):
    if i == 0:
        pc.set_facecolor('skyblue')  # Data Diff
    else:
        pc.set_facecolor('lightgreen')  # Data Same
    pc.set_edgecolor('black')
    pc.set_alpha(0.7)

axes[0].set_title('First 20 columns')
axes[0].set_xticks([1, 2])
axes[0].set_xticklabels(['Data Diff', 'Data Same'])

# 添加显著性检验
add_significance(axes[0], data_diff_avg1, data_same_avg1, x_offset=1)

# 第二组：data_diff和data_same的后20列平均值比较
parts = axes[1].violinplot([data_diff_avg2, data_same_avg2], showmedians=True, showmeans=False, showextrema=False)

# 设置颜色
for i, pc in enumerate(parts['bodies']):
    if i == 0:
        pc.set_facecolor('skyblue')  # Data Diff
    else:
        pc.set_facecolor('lightgreen')  # Data Same
    pc.set_edgecolor('black')
    pc.set_alpha(0.7)

axes[1].set_title('Last 20 columns')
axes[1].set_xticks([1, 2])
axes[1].set_xticklabels(['Data Diff', 'Data Same'])

# 添加显著性检验
add_significance(axes[1], data_diff_avg2, data_same_avg2, x_offset=1)

plt.tight_layout()
plt.show()
```
